### Solução para o exercício 2

#### Código correspondente ao arquivo Exercise2.cs

```python

namespace AT
{
    public class AT_02
    {
        public void Exercise2()
        {

            while (true)
            {
                Console.WriteLine("Digite um nome a ser Cifrado");
                string name = Console.ReadLine();

                if (!Program.ValidarEntrada(name))
                {
                    Console.WriteLine("Entrada inválida, por favor digite apenas dígitos alfabéticos!");
                    break;
                }

                char[] result = new char[name.Length];


                for (int i = 0; i < name.Length; i++)
                {
                    char c = name[i];

                    if (c == ' ')
                    {
                        result[i] = ' ';
                    }

                    else
                    {

                        if (char.IsUpper(c))
                        {
                            int newIndex = (c - 'A' + 2) % 26 + 'A';

                            result[i] = (char)newIndex;
                        }
                        else if (char.IsLower(c))
                        {
                            int newIndex = (c - 'a' + 2) % 26 + 'a';

                            result[i] = (char)newIndex;
                        }
                    }
                }

                string converted = new string(result);
                Console.WriteLine("Resultado: " + converted);
                break;
            }
        }
    }
}
```

Neste código fazemos a captura da entrada do usuário, e logo confirmamos se se trata de um nome válido sem caracteres numéricos ou especiais. String validada, seguimos para um loop for, onde a string é destrinchada a cada caracter e logo é verificado se se trata de um espaço vazio, sendo vazio, é adicionado um espaço na nova string, não sendo vazio, é confirmado se o caracter é maiúsculo ou minusculo, em mabos os casos, é utilizado a lista ASCII para busca e validação, sendo caracteres maiúsculos de 65 - 90 e minúsculos de 97 - 122. Assim o caracter é subratído do índice inicial da lista o que nos resulta em um índice de 0 - 26, logo adicionamos 2, que é a quantidade de casas que queremsos pular. Essa conta inicial é inserida em % 26, para garantir que o resultado se mantenha dentro do limite permitido do alfabeto correspondente e ao final é adicionado 'a' ou 'A' que corresponde ao primeiro índice da lista ASCII, nos resultando no valor correspondente a c adicionados de 2 casas no sistema ASCII. Esse resultado é então armazenado em um array result que é convertido para string em string converted.

```python
namespace AT
{
    public class AT_03
    {
        public void Exercise3()
        {
            while (true)
            {
                Console.WriteLine("Digite dois números separados por espaço diferentes de 0 (Formato 'XX XX')");
                string number = Console.ReadLine();
                Console.WriteLine(" ");


                // Permite a entrada de números com 0 à sua direita, mas não numerais 0 somente, também, exige separação dos númeors por espaço.
                if (!Program.validarEntradaNumericaComEspaco(number))
                {
                    Console.WriteLine("Entrada inválida!");
                    continue;
                }

                string[] splittedNumbers = number.Split(" ");

                int[] convertedNumbers = new int[splittedNumbers.Length];

                for (int i = 0; i < splittedNumbers.Length; i++)
                {
                    convertedNumbers[i] = Convert.ToInt32(splittedNumbers[i]);
                }


                int a = convertedNumbers[0];
                int b = convertedNumbers[1];

                Console.WriteLine("Digite o número da opção desejada");
                Console.WriteLine("01 - Soma");
                Console.WriteLine("02 - Subtração");
                Console.WriteLine("03 - Multiplicação");
                Console.WriteLine("04 - Divisão");
                Console.WriteLine(" ");

                string operation = Console.ReadLine();

                if (!Program.validarEntradaNumericaSemEspaco(operation))
                {
                    Console.WriteLine("Número de operação inválido!");
                    continue;
                }

                int convertedOperation = Convert.ToInt32(operation);

                switch (convertedOperation)
                {
                    case 1: Console.WriteLine($"A soma de {a} e {b} é: {a + b}");
                        break;
                    case 2: Console.WriteLine($"A subtração de {a} e {b} é: {a - b}");
                        break;
                    case 3: Console.WriteLine($"A multiplicação de {a} e {b} é: {a * b}");
                        break;
                    case 4: Console.WriteLine($"A divisão de {a} e {b} é: {a / b}");
                        break;
                    default: Console.WriteLine("Opção da operação inválida, operação cancelada");
                        break;
                }
                break;
            }
        }
    }
}
```

```python
using System;
using System.Globalization;


namespace AT
{
    public class AT_04
    {
        public void Exercise4()
        {
            while(true)
            {
                DateTime date = DateTime.Now;
                CultureInfo ptBr = CultureInfo.CreateSpecificCulture("pt-BR");

                Console.WriteLine("Digite sua data de nascimento (Formato dd/mm/YYYY)");
                string birthday = Console.ReadLine();

                if (!Program.validadeEntradaDeData(birthday))
                {
                    Console.WriteLine("Entrada inválida!");
                    continue;
                }

                DateTime convertedBirthday = DateTime.ParseExact(birthday, "dd/MM/yyyy", ptBr);


                int birthdayDay = convertedBirthday.Day;
                int birthdayMonth = convertedBirthday.Month;
                int bithdayYear = convertedBirthday.Year;

                int dateDay = date.Day;
                int dateMonth = date.Month;
                int dateYear = date.Year;

                DateTime nextBirthday;

                if (birthdayMonth < dateMonth || (birthdayMonth == dateMonth && birthdayDay < dateDay))
                {
                    nextBirthday = new DateTime(dateYear + 1, birthdayMonth, birthdayDay);
                }
                else
                {
                    nextBirthday = new DateTime(dateYear, birthdayMonth, birthdayDay);
                }

                TimeSpan intervalo = nextBirthday - date;


                if (intervalo.Days < 7)
                {
                    Console.WriteLine($"Seu aniversário está muitoooo perto, faltam {intervalo.Days} dias pra chegar.");
                    break;
                }

                Console.WriteLine($"Faltam {intervalo.Days} dias para o próximo aniversário.");
                break;
            }    
        }
    }
}
```

Neste código estamos utilizando de três classes para manipulação de data e formato, DateTime ao qual instancia uma nova data tipo DateTime, TimeSpan ao qual permite o cálculo do intervalo de datas em dias, e Globalization, ao qual permite settar o formato da data para o tipo desejado (pt-Br). Iniciamos instanciando DateTime para retornar a data atual, após instanciamos CultureInfo, uma classe de globalization, ao qual nos permite criar uma cultura específica no formato pt-Br. Logo solicitamos ao usuário sua data de nascimento, e à validamos através do método disponível em Program.cs validadeEntradaDeDAta. Sendo uma data válida convertemos para o formato pt-Br dd/mm/YYYY. Assim entramos na estrutura condicional onde é verificado se o mês de aniversário é menor que o mês atual OU se o mês é igual ao atual mas o dia de nascimento é menor que o atual, neste caso é concluído que o aniversário já passou e adicionamos 1 ao ano de dateYear (representando o ano atual).
```python
using System;
using System.Globalization;


namespace AT
{
    public class AT_05
    {
        public void Exercise5()
        {
            while (true)
            {

                CultureInfo ptBr = CultureInfo.CreateSpecificCulture("pt-BR");
                DateTime converteGraduationDate = DateTime.ParseExact("15/06/2027", "dd/MM/yyyy", ptBr);

                Console.WriteLine("Digite a data atual (formato dd/mm/YYYY");

                string date = Console.ReadLine();

                if (!Program.validadeEntradaDeData(date))
                {
                    Console.WriteLine("Entrada Inválida!");
                }

                DateTime converteDate = DateTime.ParseExact(date, "dd/MM/yyyy", ptBr);

                int graduationDay = converteGraduationDate.Day;
                int graduationMonth = converteGraduationDate.Month;
                int graduationYear = converteGraduationDate.Year;

                int dateDay = converteDate.Day;
                int dateMonth = converteDate.Month;
                int dateYear = converteDate.Year;

                int dayDiff = graduationDay - dateDay;
                int monthDiff = graduationMonth - dateMonth;
                int yearDiff = graduationYear - dateYear;

                TimeSpan interval = converteGraduationDate - converteDate;

                if (converteDate > DateTime.Now)
                {
                    Console.WriteLine("Erro: A data informada não pode ser no futuro!");
                    break;
                }

                if (interval.TotalDays < 0)
                {
                    Console.WriteLine("A data de formatura já passou!");
                    Console.WriteLine("Parabéns pela conquista!");
                    break;
                }
                else
                {
                    if (monthDiff < 0)
                    {
                        monthDiff += 12;
                        yearDiff--;
                    }

                    if (dayDiff < 0)
                    {
                        int lastMonth = converteDate.Month - 1;
                        int lastYear = converteDate.Year;

                        if (lastMonth == 0)
                        {
                            lastMonth = 12;
                            lastYear--;
                        }

                        DateTime primeiroDiaDoMesAtual = new DateTime(lastYear, lastMonth, 1);
                        DateTime ultimoDiaDoMesAnterior = primeiroDiaDoMesAtual.AddDays(-1);
                        int daysInLastMonth = ultimoDiaDoMesAnterior.Day;

                        dayDiff += daysInLastMonth;
                        monthDiff--; 
                    }
                    if (monthDiff <= 6 && yearDiff == 0)
                    {
                        Console.WriteLine($"Faltam {monthDiff} meses e {dayDiff} dias para a sua formatura!.");
                        Console.WriteLine("A reta final chegou! Prepare-se para a formatura!");
                        break;

                    }
                    Console.WriteLine($"Faltam {yearDiff} anos, {monthDiff} meses e {dayDiff} dias para a sua graduação.");
                    break;
                }
            }
        }
    }
}
```

Se a data fornecida for válida, ela é convertida de string para DateTime.

A partir daí, tanto a data de formatura quanto a data fornecida pelo usuário são separadas em dia, mês e ano. Utilizamos essas informações para calcular a diferença entre as duas datas. O tipo TimeSpan é utilizado para calcular o intervalo em dias entre a data da formatura e a data informada. Isso nos dá um valor exato do tempo restante ou se a data de formatura já passou.

O programa então entra em uma estrutura condicional para verificar algumas condições. Caso a data fornecida pelo usuário seja maior que a data atual, ou seja, se o usuário fornecer uma data no futuro, o programa exibe um erro, pois não é possível informar uma data futura. Caso a diferença entre as datas seja negativa, ou seja, se a formatura já passou, o programa celebra a conquista com uma mensagem de parabenização.

Se a diferença de meses for negativa, isso indica que o mês da data fornecida ainda não foi completado. Além disso, se a diferença de dias também for negativa, isso indica que o dia ainda não completou dentro do mês. Para corrigir, calculamos o último dia do mês anterior, levando em consideração o último mês e ano da data informada. Esse ajuste é feito subtraindo os dias necessários para completar o mês e subtraindo 1 do valor de diffMonth, já que o mês da data informada ainda não foi completado.

Por fim, se a diferença de anos for menor ou igual a 0 e a diferença de meses for inferior a 6, o programa exibe uma mensagem especial de parabenização, indicando que a formatura está próxima. ### Solução para o exercício 10

#### Código presente no arquivo Exercise10.cs

```python


namespace AT_10
{
    public class Exercise10
    {

        public void Adivinhacao()
        {
            Random random = new Random(); 
            int numeroAleatorio = random.Next(1, 51); 
            int tentativas = 5; 
            bool numeroCorreto = false; 

            Console.WriteLine("Bem-vindo ao jogo! Adivinhe o número de 1 a 50.");
            Console.WriteLine("Você tem 5 tentativas!");

            for (int i = 0; i < tentativas; i++)
            {
                try
                {
                    Console.WriteLine($"\nTentativa {i + 1} de {tentativas}:");
                    Console.Write("Digite um número: ");
                    string input = Console.ReadLine();

                    if (!int.TryParse(input, out int palpite))
                    {
                        Console.WriteLine("Por favor, digite um número válido.");
                        continue; 
                    }

                    if (palpite < 1 || palpite > 50)
                    {
                        ExibirX();
                        Console.WriteLine("Erro: O número deve estar entre 1 e 50.");
                        continue;
                    }

                    if (palpite == numeroAleatorio)
                    {
                        numeroCorreto = true;
                        Console.WriteLine($"Parabéns! Você acertou o número {numeroAleatorio}.");
                        break; 
                    }
                    else if (palpite > numeroAleatorio)
                    {
                        Console.WriteLine("O número é menor que o seu palpite.");
                    }
                    else
                    {
                        Console.WriteLine("O número é maior que o seu palpite.");
                    }
                }
                catch (Exception ex)
                {
                    Console.WriteLine($"Ocorreu um erro inesperado: {ex.Message}");
                }
            }

            if (!numeroCorreto)
            {
                Console.WriteLine($"Você perdeu! O número era {numeroAleatorio}.");
            }
        }
             
        
        public static void ExibirX()
        {
            int tamanho = 25;
            for (int i = 0; i < tamanho; i++)
            {
                for (int j = 0; j < tamanho; j++)
                {
                    if (i == j || i + j == tamanho - 1)
                    {
                        Console.Write("X");
                    }
                    else
                    {
                        Console.Write(" ");
                    }
                }
                Console.WriteLine();
            }
        }
    }
}
```

Neste programa, iniciamos por realizar as devidas instâncias, primeiro de Random, para obter o número aleatório entre 1 e 50, logo as tentativas são setadas para 5 como número máximo de tentativas, o uma bandeira setada para false, para acompanhar nosso jogador.

O programa segue dando boas vindas e informando que o jogador pode fazer o primeiro palpite, logo um loop for é iniciado com i sendo menor que tentativas. Neste momento é informado que o jogador está na tentativa i +1 (Já alterando o valor de i), A captura do palpite é efetuada e verificada, utilizando o método disponível em Program.cs e a conversão via Converto.ToInt, não sendo possível é então exibido um erro, sendo possível é feito a verificação. Se o palpite estiver fora dos limites, a função ExibirX é chamada onde é exibido um "X" grande na tela, dizendo que ele precisa se manter no limite de 1 e 50. Se o palpite for o número secreto, o programa parabeniza e encerra, se for menor indica que precisa chutar mais alto e vise versa, até encerrar as tentativas e programa encerrar por falta de tentativas sobrando.

##### FormatoModifier.cs précria as classes ContatoFormatter e ExibirContatos.

```python
namespace AT
{
    public abstract class ContatoFormatter
    {
        public abstract void ExibirContatos(List<Contato> contatos);
    }

}
```

##### MarckDown.cs - Exibição em Marckdown

```python

namespace AT
{
    public class MarkdownFormatter : ContatoFormatter
    {
        public override void ExibirContatos(List<Contato> contatos)
        {
            
            foreach (var contato in contatos)
            {
                Console.WriteLine($"- Nome: {contato.Nome}");
                Console.WriteLine($"- Telefone: {contato.Telefone}");
                Console.WriteLine($"- Email: {contato.Email}");
                Console.WriteLine();
            }
        }
    }
}
```

##### Raw.text.cs - Exibição em RawText

```python
namespace AT
{
    public class RawTextFormatter : ContatoFormatter
    {
        public override void ExibirContatos(List<Contato> contatos)
        {
            foreach (var contato in contatos)
            {
                Console.WriteLine($"Nome: {contato.Nome} | Telefone: {contato.Telefone} | Email: {contato.Email}");
            }
        }
    }

}
```

##### TabelaFormat.cs - Exibição em Tabela

```python

namespace AT
{
    public class TabelaFormatter : ContatoFormatter
    {
        public override void ExibirContatos(List<Contato> contatos)
        {
            
            foreach (var contato in contatos)
            {
                Console.WriteLine($"| {contato.Nome} | {contato.Telefone} | {contato.Email} |");
            }

            Console.WriteLine("----------------------------------------");
        }
    }
}

```
